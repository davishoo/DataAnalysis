```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import  matplotlib
myfont = matplotlib.font_manager.FontProperties(fname='/Library/Fonts/Songti.ttc')
df = pd.DataFrame(np.random.rand(6, 4),
                    index=['如心南湾', '英皇骏景', '中环荷里活', 'TUVE', '港岛太平洋', '城市花园'],
                    columns=pd.Index(['酒店销售', '携程', '艺龙', '马蜂窝'], name='湾仔秋季电子展'))
```

```python
%matplotlib inline
```

```python
fig = plt.figure()
ax1 = plt.subplot(2, 1, 1)

df.plot.bar(ax=ax1, sharex=True, figsize=(10,8)) # refer to book "DataAnalysis"

plt.title(u"香港湾仔秋季电子展", fontproperties = myfont) # 标题
plt.xlabel(u"酒店", fontproperties = myfont)
plt.ylabel(u"价格", fontproperties = myfont)
plt.xticks(np.arange(len(df.index)), df.index, fontproperties = myfont, rotation=360)
plt.legend(prop = myfont)


ax2 = plt.subplot(2, 1, 2)

df.plot.bar(ax=ax2, sharex=True, figsize=(10,8)) # refer to book "DataAnalysis"

plt.title(u"香港湾仔秋季电子展", fontproperties = myfont) # 标题
plt.xlabel(u"酒店", fontproperties = myfont)
plt.ylabel(u"价格", fontproperties = myfont)
plt.xticks(np.arange(len(df.index)), df.index, fontproperties = myfont, rotation=0)
plt.legend(prop = myfont)
```
